# 추가 데이터 세트
- 임의로 확실한 노이즈가 포함되어 있는 데이터 세트에 대해서 실험
- 실제 TrnaAD with noise 모델이 효과가 있는지 확인하기
- 특정 분포가 정해져 있는 데이터 세트에 대해서 노이즈를 추가하여, 이 노이즈를 제대로 식별하는지...? 혹은 이상치 식별 비율을 제대로 확인하는 지 확인
- 근데 애초부터 클러스터링 용 모델이 아니고, 해당 분포에서 rare한 데이터 포인트에 대해서 식별하는게 목표인 모델인데, 어떻게 노이즈를 추가해야하는 거지. 애초에 시계열 데이터가 아닌 데이터 세트에 대해서는 노이즈라른 것이.. 어떤 소용이 있는걸까

```python
import numpy as np
import pandas as pd

# 1. 데이터 설정
num_samples = 10000  # 총 샘플 수
num_features = 5    # 특성 수
imbalance_ratio = 0.95  # 클래스 불균형 비율 (0.9:0.1)

# 2. 클래스 레이블 생성
num_class_0 = int(num_samples * imbalance_ratio)  # 클래스 0의 샘플 수
num_class_1 = num_samples - num_class_0          # 클래스 1의 샘플 수

# 클래스 0과 1의 레이블 생성
class_0_labels = np.zeros(num_class_0)
class_1_labels = np.ones(num_class_1)
labels = np.concatenate([class_0_labels, class_1_labels])

# 레이블을 섞어서 시계열 데이터의 랜덤성을 보장
np.random.shuffle(labels)

# 3. 시계열 특성 데이터 생성
# 임의의 특성 값을 생성 (정규분포 또는 랜덤한 패턴)
features = np.random.randn(num_samples, num_features)

# 시계열 순서를 추가 (시간축)
time_steps = np.arange(num_samples)

# 4. 데이터프레임 생성
data = pd.DataFrame(features, columns=[f'Feature_{i+1}' for i in range(num_features)])
# data['Time'] = time_steps
data['Label'] = labels

# 5. 데이터 확인
print(data.head())
```

```python
# 트렌드 추가
# trend = 0.05 * time_steps  # 시간에 따라 증가하는 값
# features_with_trend = features + trend.reshape(-1, 1)

# 노이즈 추가
noise = np.random.normal(0, 0.1, size=(num_samples, num_features))
features_realistic = features + noise

data.iloc[:, :-1] = features_realistic  # 특성 업데이트
```

```python
1. npy로 만들기
2. train데이터셋은 label이 0인 샘플 중 4000개를 할당
3. test 데이터셋은 그 나머지 샘플로 하고, 레이블 파일을 따로 만들기 
즉, train, test, labels파일로 만들어두기 두고 이미 존재하는 폴더인 processed에 My폴더를 생성해 .npy파일로 저장해두기 
```

```python
import os
import numpy as np
import pandas as pd

# 데이터프레임 예제 (이 코드는 실제 데이터로 대체되어야 합니다)
# data = pd.DataFrame({ 'Feature_1': [...], 'Feature_2': [...], 'Label': [...] })
# 아래는 테스트용 예제 데이터
num_samples = 10000  # 전체 샘플 수
num_features = 5     # 특성 개수

# 임의 데이터 생성 (0과 1의 레이블로 구성)
data = pd.DataFrame(
    np.random.randn(num_samples, num_features), 
    columns=[f'Feature_{i+1}' for i in range(num_features)]
)
data['Label'] = np.concatenate([np.zeros(9000), np.ones(1000)])  # 레이블 9:1 비율
data = data.sample(frac=1).reset_index(drop=True)  # 셔플

# **1. Train/Test 데이터셋 분리**
# 레이블이 0인 샘플 선택
label_0_samples = data[data['Label'] == 0]

# Train 데이터셋으로 4000개 할당
train_data = label_0_samples.iloc[:4000]

# Test 데이터셋으로 나머지 샘플
test_data = data.iloc[~data.index.isin(train_data.index)]

# Train, Test 레이블 생성
train_labels = train_data['Label']
test_labels = test_data['Label']

# Label 열 제외한 데이터 (Train/Test)
train_data = train_data.drop(columns=['Label'])
test_data = test_data.drop(columns=['Label'])

# **2. 폴더 생성 및 저장**
# My 폴더 경로 설정
output_folder = '0.Nosing_effect/TranAD/processed/My'
os.makedirs(output_folder, exist_ok=True)  # 폴더 생성 (이미 존재해도 문제없음)

# Train, Test, Labels 데이터 저장
np.save(os.path.join(output_folder, 'train.npy'), train_data.to_numpy())
np.save(os.path.join(output_folder, 'test.npy'), test_data.to_numpy())
np.save(os.path.join(output_folder, 'labels.npy'), {'train': train_labels.to_numpy(), 'test': test_labels.to_numpy()})

print(f"Files saved in: {output_folder}")
```
